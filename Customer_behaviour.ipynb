```python
import pandas as pd
df = pd.read_csv('/content/customer_shopping_behavior.csv')
```

```python
df.isnull().sum()
```

```python
# imp --> i have filled the null values with the median values of the items in that particular category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.head()
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
# new columns based on age group
labels = ['Young Adults', 'Adults', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# creating new column for purchase frequcny days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months': 90,
    'Quarterly' : 90
}

```

```python
df['purchase_frequency_days'] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[['purchase_frequency_days',"frequency_of_purchases"]].head(10)
```

```python
df[['discount_applied','promo_code_used']].head()
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis = 1)
```
